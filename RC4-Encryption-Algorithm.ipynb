```
import sys

key = input("Enter the key : ")
ptext = input("Enter the message you want to encrypt : ")

key = [ord(letter) for letter in key]

def KSA(key):  
  keylen = len(key)
  state = [i for i in range(256)]

  j = 0
  for i in range(len(state)):
    j = (j + state[i]  + key[i % keylen]) % 256
    state[i], state[j] = state[j], state[i]
  return state

def PRGA(state):
  i = 0
  j = 0
  while True:
    i = (i + 1) % 256
    j = (j + state[i]) % 256
    state[i], state[j] = state[j], state[i]
    key = state[(state[i] + state[j]) % 256]
    yield key

state = KSA(key)
stream = PRGA(state)

for letter in ptext:
  sys.stdout.write("%02X" % (ord(letter) ^ stream.__next__()))
```
